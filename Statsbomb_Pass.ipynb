```python
event = events[['team','type','minute','location','pass_end_location','player','pass_recipient','pass_length','pass_outcome','pass_height']]
```

```python
liverpool_pass = event[(event['team'] == 'Liverpool') & (event['type'] == 'Pass')].sort_values('minute').reset_index()
liverpool_pass
```

```python
# liverpool_pass.drop('index',axis = 1,inplace=True)
```

#Pass map

```python
liverpool_pass['pass_outcome'].unique()
```

```python
liverpool_pass['pass_outcome'] = liverpool_pass['pass_outcome'].fillna('Success')
```

```python
liverpool_pass[['x','y']] = liverpool_pass['location'].apply(pd.Series)
liverpool_pass[['end_x','end_y']] = liverpool_pass['pass_end_location'].apply(pd.Series)
```

```python
# liverpool_pass.drop(['location','pass_end_location'],axis=1,inplace = True)
```

```python
liverpool_pass
```

```python
liverpool_pass['player'].unique()
```

```python
import matplotlib.pyplot as plt
from mplsoccer.pitch import Pitch
```

```python
# player_pass['pass_height'].unique()
```

```python
fig, ax = plt.subplots(figsize=(13.5,8))
fig.set_facecolor('#22312b')
ax.patch.set_facecolor('#22312b')
pitch = Pitch(pitch_type='statsbomb', orientation='horizontal',
              pitch_color='#22312b', line_color='#c7d5cc', figsize=(16, 11),
              constrained_layout=True, tight_layout=True)
pitch.draw(ax =ax)
plt.gca().invert_yaxis()

player_pass = liverpool_pass[(liverpool_pass['player'] == 'James Philip Milner') & (liverpool_pass['minute'] < 91)].reset_index()

for i in range(len(player_pass)):
  if player_pass['pass_outcome'][i] == 'Success':
    ax.plot((player_pass['x'][i],player_pass['end_x'][i]),(player_pass['y'][i],player_pass['end_y'][i]),color = 'green')
    ax.scatter(player_pass['x'][i],player_pass['y'][i],color ='green')
  if (player_pass['pass_outcome'][i] == 'Success') & (player_pass['pass_height'][i] == 'High Pass'):
    ax.plot((player_pass['x'][i],player_pass['end_x'][i]),(player_pass['y'][i],player_pass['end_y'][i]),ls = '--',color = 'green')
    ax.scatter(player_pass['x'][i],player_pass['y'][i],color ='green')
  if player_pass['pass_outcome'][i] == 'Incomplete':
    ax.plot((player_pass['x'][i],player_pass['end_x'][i]),(player_pass['y'][i],player_pass['end_y'][i]),color = 'red')
    ax.scatter(player_pass['x'][i],player_pass['y'][i],color ='red')
```

#Heatmap of a player in the match

```python
pitch = Pitch(orientation='vertical',pitch_color='#22312b', line_color='#c7d5cc', figsize=(16, 11))
fig,ax = pitch.draw()
fig.set_facecolor('#22312b')
sns.kdeplot(p1['x'],p1['y'],cmap='coolwarm',shade=True,shade_lowest=False,alpha = 0.7,thresh=0.4,levels=100,vertical=True,linewidth=0)
plt.xlim(0,80)
plt.ylim(0,120)
```
